```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ratings = pd.read_csv('player_ratings.csv')[['Player', 'href', 'Year', 'G', 'GS', 'MP', 'PTS', 'PER', 'WS', 'BPM', 'VORP']]
ratings['VORP'] = ratings['VORP'] / ratings['G']
ratings['WS'] = ratings['WS'] / ratings['G']
ratings = ratings[(ratings['G'] >= 5) & (ratings['MP'] >= 1) & (ratings['G'] *  ratings['MP'] > 50)]
```

```python
def PlayerStrength(player, year) :
    df = ratings.copy()[(ratings['href'] == player) & (ratings['Year'] <= year) & (ratings['Year'] > year - 3)]
    df['weight'] = 1 / 2 ** (year - df['Year'])
    row = df.tail(1)[['Player', 'href', 'Year']]
    for f in ['PTS', 'PER', 'WS', 'BPM', 'VORP'] :
        row[f] = (df[f] * df['weight']).sum() / df['weight'].sum()
    return row
```

```python
data = [PlayerStrength(player, year) for player, year in zip(ratings['href'], ratings['Year'])]
```

```python
data = pd.concat(data)
data.head()
```

```python
q = data.copy()
for f in ['PTS', 'PER', 'WS', 'BPM', 'VORP'] :
    q[f] = q[f].apply(lambda x: (x >= data[f]).sum() / len(data))
```

```python
stats = q[['PTS', 'PER', 'WS', 'BPM', 'VORP']]
q['q'] = stats.mean(axis = 1)
```
